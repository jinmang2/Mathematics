# 1. 함수와 극한

$Def.$ function $f$는 집합 $A$에 속하는 각 원소 $x$에 집합 $B$에 속하고 $f(x)$로 나타내는 단 한 개의 원소를 대응시키는 규칙

$note\;that.$
- domain: 정의역
- range: 치역
- value of $f$ at $x$: $x$에서의 $f$의 값

$vertical\;line$ 판정법:
- $Def.$ Let X and Y be sets. A **function** from X to Y is a triple (f, X, Y), where f is a relation from X to Y satisfying
    1. Dom(f) = X
    2. If (x, y) $\in$ f and (x, z) $\in$ f, then y=z
- 2번 조건은 f(x)=y and f(x)=z이면 y=z라는 얘기
- 즉, y=z가 성립이 안되면 함수가 아니다.
- 이를 활용한 함수 판별법

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# adjust 한글 font
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)

plt.rcParams['axes.unicode_minus'] = False
```

```python
f = lambda x, a, b, c, d, e:  a * x ** 4 + b * x ** 3 + c * x ** 2 + d * x + e
g = lambda x, a, b: a * 2 ** x + b
plt.figure(figsize=(10, 8), facecolor='w')
head = np.linspace(0, 10, 100)
tail = np.linspace(10, 7, 30)
ind = np.hstack((head, tail))
val = np.hstack(
    (f(head, 0.5, -4, -10, 4, 610),
     g(tail, (1/16), -14+600)))
plt.axvline(0, color='k')
plt.axhline(0, color='k')
plt.axvline(8.5, color='r', lw=3)
pd.Series(index=ind, data=val).plot(grid=True, lw=3)
plt.title('함수가 아닌 예(vertical line 판정법)', fontsize=18)
plt.xticks(range(-1, 12)); plt.show()
```

### Odd / Even Function
- `Even Function`: $f(-x)\;=\;f(x)$
- `Odd Function`: $f(-x)\;=\;-f(x)$

```python
# Even Function
x = np.linspace(-100, 100, 201)
y = x ** 2 - 1000
plt.figure(figsize=(10, 8), facecolor='w')
plt.plot(x, y, lw=3)
plt.grid(True)
plt.axvline(0, color='k')
plt.axhline(0, color='k')
plt.title('Even Function', fontsize=18)
plt.show()
```

```python
# Odd Function
x = np.linspace(-100, 100, 201)
y = x ** 3
plt.figure(figsize=(10, 8), facecolor='w')
plt.plot(x, y, lw=3)
plt.grid(True)
plt.axvline(0, color='k')
plt.axhline(0, color='k')
plt.title('Odd Function', fontsize=18)
plt.show()
```

### Power function
$f(x)=x^a$

```python
# a = n such that n > 0
x = np.linspace(-100, 100, 201)
fig, ax = plt.subplots(nrows=1, ncols=5, figsize=(10*5, 8), facecolor='w')
def plot_power_func(ax, x, pow_):
    ax.plot(x, x ** pow_, lw=3)
    ax.grid(True)
    ax.axvline(0, color='k')
    ax.axhline(0, color='k')
    ax.set_title(f'Power Function; n={pow_}', fontsize=18)

for i in range(1, 6):
    plot_power_func(ax[i-1], x, i)

plt.show()
```

```python
# a = 1 / n such that n > 0
x = np.linspace(-100, 100, 201)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10*2, 8), facecolor='w')

for i, pow_ in zip(range(1, 3), [1/2, 1/3]):
    plot_power_func(ax[i-1], x, pow_)

plt.show()
# x ** (1/3)은 odd function
# 지금 잘못 나와있는 것
```

```python
# reciprocal function
x = np.linspace(-10, 10, 201)
plt.figure(figsize=(10, 8), facecolor='w')
ax = plt.gca()
plot_power_func(ax, x, -1)
```

### Trigonometric function
#### sin function
$f(x) = sin(x)$
#### cos function
$f(x) = cos(x)$
#### tan function
$f(x) = tan(x) = \cfrac{sin(x)}{cos(x)}$

```python
# a = n such that n > 0
x = np.linspace(-10, 10, 2000)
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10*3, 8), facecolor='w')
def plot_trigonometric_func(ax, x, mode):
    if mode == 'sin':
        y = np.sin(x)
    elif mode == 'cos':
        y = np.cos(x)
    elif mode == 'tan':
        y = np.tan(x)
    else:
        raise ValueError('sin, cos, tan 중에 하나의 함수를 넣어주세요.')
    ax.plot(x, y, lw=3)
    ax.grid(True)
    ax.axvline(0, color='k')
    ax.axhline(0, color='k')
    mode = mode.capitalize()
    ax.set_title(f'{mode} Function', fontsize=18)

for i, mode in enumerate(['sin', 'cos', 'tan']):
    plot_trigonometric_func(ax[i], x, mode)

plt.show()
```

### Exponential & Log function
- Exponential function
$f(x)=a^x$
- log function
$f(x)=log_a{x}$

```python
# a = n such that n > 0
x = np.linspace(-10, 10, 2000)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10*2, 8), facecolor='w')
def plot_exp_log_func(ax, x, mode):
    if mode == 'exp':
        y = np.exp(x)
    elif mode == 'log':
        y = np.log(x)
    else:
        raise ValueError('exp, log 중에 하나의 함수를 넣어주세요.')
    ax.plot(x, y, lw=3)
    ax.grid(True)
    ax.axvline(0, color='k')
    ax.axhline(0, color='k')
    mode = mode.capitalize()
    ax.set_title(f'{mode} Function', fontsize=18)

for i, mode in enumerate(['exp', 'log']):
    plot_exp_log_func(ax[i], x, mode)

plt.show()
```

## Limit

## 1. 직관적인 정의
- $Def.$ a와 같지 않지만 a에 충분히 가까운 x의 값을 (a의 어느 쪽에서든) 택함으로써 f(x)의 값을 L에 한없이 가깝게 할 수 있을 때, "x가 a에 한없이 가까워질 때 f(x)의 극한은 L이다."라고 하고 이를 기호로 다음과 같이 나타낸다.
$$\lim_{x\to{a}}f(x)=L$$

### Q1)
$$\lim_{x\to{1}}\cfrac{x-1}{x^2-1}$$의 값을 어림하라.

```python
x_left = np.linspace(0.995, 0.9999, 100)
x_right = np.linspace(1.0001, 1.005, 100)
func = lambda x: (x - 1) / (x**2 - 1)

plt.figure(figsize=(10, 8), facecolor='w')
plt.plot(np.hstack((x_left, x_right)),
         np.hstack((func(x_left), func(x_right))), lw=0, marker='s', ms=1)
plt.axvline(1, color='r')
plt.title('0.5로 수렴하고 있음을 직관적으로 확인할 수 있다.', fontsize=18)
plt.show()
```

### Q1)
$$\lim_{x\to{0}}\cfrac{sin(x)}{x}$$의 값을 어림하라.

```python
x_left = np.linspace(-1, -0.0001, 100)
x_right = np.linspace(0.0001, 1, 100)
func = lambda x: np.sin(x) / x

plt.figure(figsize=(10, 8), facecolor='w')
plt.plot(np.hstack((x_left, x_right)),
         np.hstack((func(x_left), func(x_right))), lw=0, marker='s', ms=1)
plt.axvline(0, color='r')
plt.title('1.0로 수렴하고 있음을 직관적으로 확인할 수 있다.', fontsize=18)
plt.show()
```

### 컴퓨터의 대수 체계(Computer Algebra Systems, CAS)
- CAS에는 극한을 셈하는 명령어가 존재
- 위의 예제에서는 수렴하는 예제를 수행했으나, 실제로 진동하거나 한 값으로 수렴하지 않는 경우가 존재한다
- CAS는 수치 실험으로 극한을 찾지 않는다.
- 대신, 무한 급수 셈과 같은 좀 더 정교한 기법을 사용한다.|

### Heaviside function
$H(t)={\begin{cases}
        0\quad(t<0)\\
        1\quad(t\geq0)\\
        \end{cases}}$
- 극한을 정의할 수 없음
- `one-sided limit` 정의

$Def.\;\lim_{x\to{a}}f(x)=L{\iff}\lim_{x\to{a^-}}f(x)=L\;and\;\lim_{x\to{a^+}}f(x)=L$ 

## 극한에 대한 명확한 정의! ($\epsilon-\delta$논법)
- $Def.\;\lim_{x\to{a}}f(x)=L{\iff}Given\;\epsilon>0,\;\exists\;\delta>0,\;{\forall}x>0\;s.t\;0<|x-a|<\delta\rightarrow|f(x)-L|<\epsilon$
- 임의의 양수 $\epsilon$에 대하여 적당한 양수 $\delta$가 존재하여 $0<|x-a|<\delta$인 모든 $x$에 대하여 $|f(x)-L|<\epsilon$을 만족한다.

## 13년도 1학년때 실제 과제
- $\lim_{x\to3}x^2=9$가 수렴함을 증명하라.

### Proof.
- $Let\;a=3, f(x)=x^2, L=9$
- $Given\;\epsilon>0,$
- $\text{Purpose: }\exists\;\delta>0\;s.t\;0<|x-3|<\delta$.
- $|f(x)-L|=|x^2-9|=|x+3|\times|x-3|$
- $\text{By Triangle inequality, }|x+3|=|x-3+6|\leq|x-3|+6$
- $\text{Suppose }\delta\leq{k}\quad{s.t}\;k>0$
- $\text{Take }\delta=min(\cfrac{\epsilon}{k+6},\;k)$
- $\text{Then, }0<|x-a|<\delta\rightarrow|f(x)-L|<\epsilon$
